# Import Packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from collections import Counter
import json

import math
import statistics
```

# Read Dataset Information

```python
INFO = []
for line in open('/homedir01/eeldeeb19/Oulu Data/devices-1613042553.json', 'r'):
    INFO.append(json.loads(line))
```

```python
print(len(INFO))
```

```python
# Check
print(INFO)
```

We can see that there are  different devices in the documentation. Each device has a corresponding unique ID, type, floor, status, time of installation, location, position, and responsible.

Next, we will visualize the corresponding information for each device.

```python
# Devices Types analysis
INFO_Type = [i["type"] for i in INFO]
# print(INFO_Type)
Uniqe_INFO_Type = dict(Counter(INFO_Type))
# len(Uniqe_INFO_Type)
Key_INFO_Type = Uniqe_INFO_Type.keys()
Rep_INFO_Type = Uniqe_INFO_Type.values()

plt.stem(Key_INFO_Type,Rep_INFO_Type)
plt.xlabel('Devices Types')
plt.ylabel('Type Repitions')
# plt.savefig('Devices_Types.jpg')
plt.show()
```

From the previous figure, we can notice that the majourity of devices are CO2 devices, whearas the minority of devies are soil moisture devices

```python
# Devices Floors analysis
INFO_Floor = [i["floor"] for i in INFO]
# print(INFO_Floor)
Uniqe_INFO_Floor = dict(Counter(INFO_Floor))
# len(Uniqe_INFO_Floor)
Key_INFO_Floor = Uniqe_INFO_Floor.keys()
Rep_INFO_Floor = Uniqe_INFO_Floor.values()

plt.stem(Key_INFO_Floor,Rep_INFO_Floor)
plt.xlabel('Devices Floors')
plt.ylabel('Floor Repitions')
# plt.savefig('Devices_Floors.jpg')
plt.show()
```

From the previous figure, we can notice that the majourity of devices are installed in the first floor in (different buildings).

```python
# Devices Status analysis
INFO_status = [i["status"] for i in INFO]
# print(INFO_status)
Uniqe_INFO_status = dict(Counter(INFO_status))
# len(Uniqe_INFO_status)
Key_INFO_status = Uniqe_INFO_status.keys()
Rep_INFO_status = Uniqe_INFO_status.values()

plt.stem(Key_INFO_status,Rep_INFO_status)
plt.xlabel('Devices Status')
plt.ylabel('Status Repitions')
# plt.savefig('Devices_Status.jpg')
plt.show()
```

From the previous figure, we can notice that all devices have been already installed except one device.

```python
# Devices Installation Time analysis
INFO_Inst_Time = [i["installed"] for i in INFO]
# print(INFO_Inst_Time)
Uniqe_INFO_Inst_Time = dict(Counter(INFO_Inst_Time))
# len(Uniqe_INFO_Inst_Time)
Key_INFO_Inst_Time = Uniqe_INFO_Inst_Time.keys()
Rep_INFO_Inst_Time = Uniqe_INFO_Inst_Time.values()

plt.stem(Key_INFO_Inst_Time,Rep_INFO_Inst_Time)
plt.xlabel('Devices Installation Time')
plt.ylabel('Installation Time Repitions')
# plt.savefig('Devices_Inst_Time.jpg')
plt.show()
```

From the previous figure, we can notice that all installed devies have been installed on 06.06.2020 at 00.00.00.

```python
# Devices Location analysis
INFO_Location = [i["desc"] for i in INFO]
# print(INFO_Location)
Uniqe_INFO_Location = dict(Counter(INFO_Location))
# len(Uniqe_INFO_Location)
Key_INFO_Location = Uniqe_INFO_Location.keys()
Rep_INFO_Location = Uniqe_INFO_Location.values()

plt.stem(Key_INFO_Location,Rep_INFO_Location)
plt.xlabel('Devices Locations')
plt.ylabel('Locations Repitions')
# plt.savefig('Devices_Locations.jpg')
plt.show()
```

From the previous figure, we can notice that there are different locations for devices.

```python
# Devices Position analysis
INFO_Position_pre = [i["location"] for i in INFO]
# print(INFO_Position_pre)
INFO_Position = []

# Function to handle lists
for i in range(0,len(INFO_Position_pre)):
    j = 0
    INFO_Position.append(INFO_Position_pre[i][j]+ ',' + INFO_Position_pre[i][j+1])
#print(INFO_Position)

Uniqe_INFO_Position = dict(Counter(INFO_Position))
# len(Uniqe_INFO_Position)
Key_INFO_Position = Uniqe_INFO_Position.keys()
Rep_INFO_Position = Uniqe_INFO_Position.values()

plt.stem(Key_INFO_Position,Rep_INFO_Position)
plt.xlabel('Devices Positions')
plt.ylabel('Positions Repitions')
# plt.savefig('Devices_Positions.jpg')
plt.show()
```

From the previous figure, we can notice that devices have different positions on their locations.

```python
# Devices Responsible analysis
INFO_Responsible = [i["addedBy"] for i in INFO]
# print(INFO_Responsible)
Uniqe_INFO_Responsible = dict(Counter(INFO_Responsible))
# len(Uniqe_INFO_Location)
Key_INFO_Responsible = Uniqe_INFO_Responsible.keys()
Rep_INFO_Responsible = Uniqe_INFO_Responsible.values()

plt.stem(Key_INFO_Responsible,Rep_INFO_Responsible)
plt.xlabel('Devices Responsibles')
plt.ylabel('Responsibles Repitions')
# plt.savefig('Devices_Responsibles.jpg')
plt.show()
```
